```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("/Users/msml/Vamsi/SF/MLBasics-master/Data/SalaryData.csv")
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
m = 0
c = 0
L = 0.000000001  
n = float(len(X))
for i in range(1000): 
    Y_pred = m*X + c
    z=1/(1 + np.exp(-Y_pred))
    D_m = (-1/n) * sum(X * (z-Y))  
    D_c = (-1/n) * sum(z-Y)  
    m = m - L * D_m  
    c = c - L * D_c
print (m, c)
Y_pred = m*X + c
print(Y_pred)
z=1/(1 + np.exp(-Y_pred))
print(z)
y_cls = [1 if i>0.1 else 0 for i in z ]
print(np.array(y_cls))
plt.scatter(X,y_cls) 
plt.plot(X,y_cls)
plt.xlabel('Experience')
plt.ylabel('Sigmoidx')
plt.show()
```
